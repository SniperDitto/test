```python
import requests as req
from bs4 import BeautifulSoup as bs
res = req.get("https://finance.naver.com/sise/dividend_list.naver?sosok=KOSPI")
res
```

```python
res.text
```

```python
soup = bs(res.text,'html.parser')
soup
```

```python
title_text_arr = soup.find_all(attrs={'scope':'col'})
title_txt =''
for txt in title_text_arr:
    title_txt += txt.get_text().strip()
    title_txt += "/"
print(title_txt)
```
